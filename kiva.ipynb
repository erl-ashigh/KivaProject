```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#data extraction 
df1 = pd.read_csv('/Volumes/SanDisk/Data Career info/KIVA/kiva_data.csv')
df1.head(3)
```

# Bar Charts

```python
#Create bar charts
import matplotlib.ticker as mtick
f, ax=plt.subplots(figsize=(15,10))
sns.set_palette("Accent")
sns.barplot(x=df1['country'], y=df1['loan_amount'])

#format y axis
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)
```

```python
#hue by gender
f, ax = plt.subplots(figsize=(15, 10))

sns.barplot(data=df1, x="country", y="loan_amount", hue="gender")

fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)
```

Above charts show

On average, male recipients receive larger loans from Kiva.
There is little gender gap in El Savador. However, the above charts show that female receive less loan amont.  
Cambodia gets the largest loan amount and Philippines gets the least loan amount.

# Box Plots 
to figure to compare the distribution of loans by country.

```python
f, ax = plt.subplots(figsize=(15,10))
sns.set_style('whitegrid')
sns.set_palette('Set2')
sns.set_context('notebook', font_scale=1.25)
sns.boxplot(x='country', y='loan_amount', data=df1)
plt.show()
```

Based on the above charts, Kenya has the widest distribution. Cambodia would most likely to receive the largest loan amount.

# Violin Plots

```python
plt.figure(figsize=(16,10))
sns.set_style('darkgrid')
sns.set_palette('Set3')
sns.violinplot(x='activity', y='loan_amount', data=df1)
plt.show()
```

```python
plt.figure(figsize=(16,10))
sns.set_palette('Dark2')
sns.set_style('white')
sns.violinplot(x='country', y='loan_amount', data=df1)
```

```python
plt.figure(figsize=(16,10))
sns.set_palette('Spectral')
sns.set_style('dark')
sns.violinplot(data=df1, x='country', y='loan_amount', hue='gender', split=True)
plt.show()
```

The above charts show that women tend to recieve lower loan amount than men especially in Pakistan and Cambodia.
There is a less gender gap in El Savador; fairly similar distribution. 
Camobdia has the widest distribution.
